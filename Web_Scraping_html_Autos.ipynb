# Identificando la información en el HTML

```python
from urllib.request import urlopen, urlretrieve
from bs4 import BeautifulSoup
import pandas as pd
```

#Extraer el numero de paginas totales:

```python
html_autos = "https://carros.tucarro.com.co/atlantico/_all*payment*methods*discount_cash*discount"

response=urlopen(html_autos)
html=response.read()
soup=BeautifulSoup(html, 'html.parser')
numero_pg = soup.find('li', {'class': 'andes-pagination__page-count'})
pg=(numero_pg.getText())
num = int(pg.split()[1])
num
```

#Guardar las paginas en pages=[ ]:
adicionando pagina principal y luego las otras siguiendo el algoritmo 48*n+1:

```python
pages = []

html_autos= 'https://carros.tucarro.com.co/atlantico/_all*payment*methods*discount_cash*discount'
pages.append(html_autos)

for n in range(1,num):
    pages.append(f'https://carros.tucarro.com.co/atlantico/_Desde_{48*n+1}_NoIndex_True_all*payment*methods*discount_cash*discount')
pages
```

#Creando una rutina de scraping

```python
autoss=[]
models=[]
ciudads=[]
imageness=[]
prices = []
prices_00 = []
price_tags_00=[]

#Recorriendo todas las paginas del sitio:
for i,url in enumerate(pages):
  response=urlopen(url)
  html=response.read()
  soup=BeautifulSoup(html, 'html.parser')

  #Selecionando las tag necesarias:
  div_precios = soup.find_all('div', {'class': 'ui-search-item__group__element ui-search-price__part-without-link shops__items-group-details'})
  carross = soup.find_all('h2', {'class': 'ui-search-item__title shops__item-title'})
  modeloss = soup.find_all('li', {'class': 'ui-search-card-attributes__attribute'})
  precs_00 = soup.find_all('div', {'class': 'ui-search-price__second-line shops__price-second-line'})
  citys = soup.find_all('span', {'class': 'ui-search-item__group__element ui-search-item__location shops__items-group-details'})
  imgs = soup.find_all('img',{'class': 'shops__image-element'})
  

# Para cada elemento `div` con la clase `slick-track`, encuentra todos los elementos `span` con la clase `price-tag-fraction`
  #Seleccionando elementos con descuendo que generan doble valor:
  for track in div_precios:
    if(len(track.get_text())>30):
      price_tags_00 = track.find_all('span', {'class':'price-tag-fraction'})

  for tag in price_tags_00:
        price = tag.text.strip()
        prices_00.append(price)

  for track in div_precios:
      price_tags = track.find_all('span', {'class':'price-tag-fraction'})
      
      # Para cada elemento `span` encontrado, extrae su contenido y agrégalo a la lista de precios
      for tag in price_tags:
          price = tag.text.strip()
          prices.append(price)
  
  for h2 in carross:
    autoss.append(h2.getText())

  for li in modeloss:
    models.append(li.get_text())

  for div in citys:
    ciudads.append(div.get_text())

  for img in imgs:
    imageness.append(img.get('data-src'))
  
for i in range(len(prices)):
    prices[i] = int(''.join(filter(str.isdigit, prices[i])))

# Lista de elementos de posición par: Modelo de auto
model_auto = [int(models[i]) for i in range(0, len(models), 2)]

# Lista de elementos de posición impar: kilomtetraje
kms = [models[i] for i in range(1, len(models), 2)]

#Cambiamo a int y sacamos km:
for i in range(len(kms)):
    kms[i] = int(''.join(filter(str.isdigit, kms[i])))

#Obteniendo el numero a eliminar:
eliminar= int(max(prices_00).replace(".", ""))
#Eliminando
prices.remove(eliminar)

print('total autos:',len(autoss),' ',autoss)
print('total ciudad:',len(ciudads),' ',ciudads)
print('total km:',len(kms),' ',kms)
print('total modelo:',len(model_auto),' ',model_auto)
print('total imagenes:',len(imageness),' ',imageness)
print('total prices:',len(prices),' ',prices)
```

#Creando DataFrame:

```python
#Pasar precio a precio/millones:
prices_mll = [num / 1000000 for num in prices]

#Dicionario para hacer las columnas del dataframe:
data = {'auto': autoss, 'Model': model_auto,'Kms':kms, 'Precio/millon': prices_mll, 'Ciudad': ciudads,'URL_img':imageness}

#DAtaFrame:
df_autos = pd.DataFrame(data)
```
